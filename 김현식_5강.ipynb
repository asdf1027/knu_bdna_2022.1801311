```python
df=pd.DataFrame(np.random.randn(5,5))
df.columns=['A','B','C','D','E']
df.iloc[1,1]=np.nan
df.iloc[2,2]=np.nan
df
# 없는 값에는 Nan이 출력됨
```

```python
df.isnull()
# Nan은 True 값이 출력됨
```

```python
df['B'].isnull()
```

```python
df['D'].isnull()
```

```python
df[df['C'].isnull()]
# C행에 대한 출력값
```

```python
df.isnull().sum()
# 각 행에 대한 True 갯수
```

```python
df['E'].isnull().sum()
# E행에 대한 True의 갯수
```

```python
df.dropna()
# Nan이 있는 행은 삭제됨
```

```python
df.dropna(inplace=True)
df
# df만 출력했을 경우 Nan이 있는 행도 같이 출력되지만 inplace=True를 넣으면 
# 제외된 행만 출력됨
```

```python
df=pd.DataFrame(np.random.randn(5,5))
df.columns=['A','B','C','D','E']
df.iloc[1,1]=np.nan
df.iloc[2,2]=np.nan
df
```

```python
df.dropna(subset=['C'],axis=0,inplace=True)
df
# C열에서 Nan이 있는 행이 사라짐
# axis=0은 행 삭제를 의미함
```

```python
df=pd.DataFrame(np.random.randn(5,5))
df.columns=['A','B','C','D','E']
df.iloc[1,1]=np.nan
df.iloc[2,2]=np.nan
df
```

```python
df.dropna(axis=1,inplace=True)
df
# Nan이 있는 B,C열이 삭제됨
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.countplot(x="day",data=tips)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='scatter')
plt.show()
# jointplot은 각각 data의 히스토그램도 출력함
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='kde')
plt.show()
# kind를 'kde'로 바꿔 출력하면 아래와 같이 출력됨
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,hue='species',kind='kde')
plt.show()
# hue='speices'와 kind를 'kde'로 바꿔 출력하면 아래와 같이 나옴
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='species',markers=['o','d','s'])
plt.show()
# hue='species',markers=['o','d','s']를 통해 각 데이터마다 색이 입혀짐
```

```python
titanic_size=titanic.pivot_table(index="class",columns="sex",aggfunc="size")
titanic_size
# pivot을 사용하여 class, sex를 구분함
```

```python
sns.heatmap(titanic_size)
plt.title("Heat map")
plt.show()
# heatmap()을 사용하여 아래처럼 색으로 구분함
```

```python
sns.heatmap(titanic_size,annot=True)
plt.title("Heat map")
plt.show()
# annot=True를 통해 박스 안에 숫자를 출력
```

```python
sns.heatmap(titanic_size,annot=True,fmt='d')
plt.title("Heat map")
plt.show()
# fmt='d'를 통해 박스 안 숫자를 정수 형태로 바꿔 출력
```

```python
sns.heatmap(titanic_size,cmap=sns.color_palette("PuBu"),annot=True,fmt='d')
plt.title("Heat Map")
plt.show()
# cmap=sns.color_palette("PuBu")를 통해 박스 색깔을 변경해 출력
```

```python
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("Tips On Each Day")
plt.show()
```

```python
sns.barplot(y="day",x="total_bill",data=tips)
plt.title("Tips On Each Day")
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
plt.title('Tips on each day')
plt.show()
# jitter가 True 일때 아래처럼 출력됨
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
plt.title('Tips on each day')
plt.show()
# jitter가 False 일때 아래처럼 출력됨
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.title('Tips on each day')
plt.show()
# swarmplot 사용
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Tips on each day')
plt.show()
# violinplot
```

```python
sns.barplot(y="day",x="total_bill",data=tips,hue='time')
plt.title("Tips On Each Day")
plt.show()
# hue 사용
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
plt.title('Tips on each day')
plt.show()
# hue 사용
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='time')
plt.title('Tips on each day')
plt.show()
# hue 사용
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='time',split=True)
plt.title('Tips on each day')
plt.show()
# split 활용
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
plt.title('Tips on each day with option hue=sex and dodge=True')
plt.show()
# dodge 활용
```
